```python
import re
import sys
import time
import datetime
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("googleplaystore.csv")
df
```

```python
# Checking the data type of the columns
df.info()
```

```python
# Exploring missing data, checking to see if any has NaN values
plt.figure(figsize=(7, 5))
sns.heatmap(df.isnull(), cmap='viridis')
df.isna().sum()
```

```python
df_drop = df.dropna()
```

```python
df_drop
```

```python
# Count number of unique values in category column
df_drop['Category'].unique()
```

```python
# App values encoding
le = preprocessing.LabelEncoder()
df_drop['App'] = le.fit_transform(df_drop['App'])
# This encoder converts the values into numeric values
```

```python
df_drop
```

```python
# Before filling null values we have to clean all non numerical values & unicode charachters 
replaces = [u'\u00AE', u'\u2013', u'\u00C3', u'\u00E3', u'\u00B3', '[', ']', "'"]
for i in replaces:
	df_drop['Current Ver'] = df_drop['Current Ver'].astype(str).apply(lambda x : x.replace(i, ''))

regex = [r'[-+|/:/;(_)@]', r'\s+', r'[A-Za-z]+']
for j in regex:
	df_drop['Current Ver'] = df_drop['Current Ver'].astype(str).apply(lambda x : re.sub(j, '0', x))

df_drop['Current Ver'] = df_drop['Current Ver'].astype(str).apply(lambda x : x.replace('.', ',',1).replace('.', '').replace(',', '.',1)).astype(float)
```

```python
# Category features encoding
category_list = df_drop['Category'].unique().tolist() 
category_list = ['cat_' + word for word in category_list]
df_drop = pd.concat([df_drop, pd.get_dummies(df_drop['Category'], prefix='cat')], axis=1)
```

```python
df_drop
```

```python
# Genres features encoding
le = preprocessing.LabelEncoder()
df_drop['Genres'] = le.fit_transform(df_drop['Genres'])
```

```python
df_drop
```

```python
# Encode Content Rating features
le = preprocessing.LabelEncoder()
df_drop['Content Rating'] = le.fit_transform(df_drop['Content Rating'])
```

```python
df_drop
```

```python
# Price cleaning
df_drop['Price'] = df_drop['Price'].apply(lambda x : x.strip('$'))
```

```python
df_drop
```

```python
# Installs cleaning
df_drop['Installs'] = df_drop['Installs'].apply(lambda x : x.strip('+').replace(',', ''))
```

```python
df_drop
```

```python
# Type encoding
df_drop['Type'] = pd.get_dummies(df_drop['Type'])
```

```python
df_drop
```

```python
# Last Updated encoding
df_drop['Last Updated'] = df_drop['Last Updated'].apply(lambda x : time.mktime(datetime.datetime.strptime(x, '%B %d, %Y').timetuple()))
```

```python
# Convert kbytes to mbytes 
k_indices = df_drop['Size'].loc[df_drop['Size'].str.contains('k')].index.tolist()
converter = pd.DataFrame(df_drop.loc[k_indices, 'Size'].apply(lambda x: x.strip('k')).astype(float).apply(lambda x: x / 1024).apply(lambda x: round(x, 3)).astype(str))
df_drop.loc[k_indices,'Size'] = converter
```

```python
# Size cleaning
df_drop['Size'] = df_drop['Size'].apply(lambda x: x.strip('M'))
df_drop[df_drop['Size'] == 'Varies with device'] = 0
df_drop['Size'] = df_drop['Size'].astype(float)
```

```python
df_drop
```

```python
df_drop[['App', 'Rating', 'Size', 'Type', 'Content Rating']].corr()
```

```python
df_drop.dtypes
```

```python
# Split data into training and testing sets
features = ['App', 'Reviews', 'Size', 'Installs', 'Type', 'Price', 'Content Rating', 'Genres', 'Last Updated', 'Current Ver']
features.extend(category_list)
X = df_drop[features]
y = df_drop['Rating']
```

```python
# Split the dataset into 75% train data and 25% test data.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 10)
```

```python
# Look at the 15 closest neighbors
model = KNeighborsRegressor(n_neighbors=15)
```

```python
# Find the mean accuracy of knn regression using X_test and y_test
model.fit(X_train, y_train)
```

```python
# Calculate the mean accuracy of the KNN model
accuracy = model.score(X_test,y_test)
'Accuracy: ' + str(np.round(accuracy*100, 2)) + '%'
```

```python
# Try different numbers of n_estimators - this will take a minute or so
n_neighbors = np.arange(1, 20, 1)
scores = []
for n in n_neighbors:
    model.set_params(n_neighbors=n)
    model.fit(X_train, y_train)
    scores.append(model.score(X_test, y_test))
plt.figure(figsize=(7, 5))
plt.title("Effect of Estimators")
plt.xlabel("Number of Neighbors K")
plt.ylabel("Score")
plt.plot(n_neighbors, scores)
```

```python
# The RandomForestRegressor

model_random_forest = RandomForestRegressor(n_jobs=-1)
# Try different numbers of n_estimators - this will take a minute or so
estimators = np.arange(10, 200, 10)
scores = []
for n in estimators:
    model_random_forest.set_params(n_estimators=n)
    model_random_forest.fit(X_train, y_train)
    scores.append(model_random_forest.score(X_test, y_test))
plt.figure(figsize=(7, 5))
plt.title("Effect of Estimators")
plt.xlabel("no. estimator")
plt.ylabel("score")
plt.plot(estimators, scores)
results = list(zip(estimators,scores))
results
```

```python
predictions = model_random_forest.predict(X_test)
'Mean Absolute Error:', metrics.mean_absolute_error(y_test, predictions)
'Mean Squared Error:', metrics.mean_squared_error(y_test, predictions)
'Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, predictions))
```
